# Palindrome Function
Se trata de una simple funcion que nos dice si una palabra es un palindromo o no.

```python

def Palindrome(word):
  palindrome = ""
  lower_word = word.lower()

  for i in range(len(word) - 1, -1, -1):
    palindrome += lower_word[i]

  if palindrome == lower_word:
    return "It's a palindrome"
  else:
    return "It's not a palindrome"
```

```python
words = ["ala", "Anilina", "arenera", "Subaru", "dama", "bob", "coc", "sexes"]

# Generamos una lista con los resultados
results = [f"{i} = {Palindrome(i)}" for i in words]

# Devolvemos la lista (Google Colab la mostrará automáticamente)
results
```

# Emordnilap Function
Se trata de una función que devuelve la palabra al reves junto a un true o un false de si es palindromo o no. Y usamos pandas para mostrar todo como una tabla

```python
def Emordnilap(word):
  emordnilap = ""
  lower_word = word.lower()

  for i in range(len(word) - 1, -1, -1):
    emordnilap += lower_word[i]

  return word, emordnilap, emordnilap == lower_word
```

```python
import pandas as pd

words = ["ala", "Anilina", "arenera", "Subaru", "dama", "bob", "coc", "sexes"]

# Desempaquetamos los valores en listas separadas
original_word, backwards_word, is_palindrome = zip(*[Emordnilap(i) for i in words])

# Mostramos los resultados
original_word, backwards_word, is_palindrome

df = pd.DataFrame({
    "Palabra Original": original_word,
    "Palabra al revés": backwards_word,
    "¿Es palindrome?": is_palindrome
})

df
```
