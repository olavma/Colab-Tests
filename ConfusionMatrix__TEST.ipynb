# Matrices de Confusi√≥n
Hay dos tipos de Matrices de Confusion, dependiendo de donde este "True Label"

Si "True Label" esta a la izquierda (donde las filas) como suele ser con Scikit-learn esta sera la matriz:

|  |  |
| --- | --- |
| TP | FN |
| FP | TN |

Por otro lado si "True Label" esta abajo (donde las columnas), un caso raro, esta sera la matriz:

|  |  |
| --- | --- |
| TP | FP |
| FN | TN |


# True Label en Filas (por defecto)

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Ejemplo de etiquetas verdaderas y predicciones
y_true = [0, 1, 0, 1, 0, 1, 1]
y_pred = [0, 0, 0, 1, 0, 1, 1]

# Generar la matriz de confusi√≥n
cm = confusion_matrix(y_true, y_pred)

# Crear la visualizaci√≥n con True Label en las filas
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap='Blues')
plt.title('Matriz de Confusi√≥n (True Label en filas)')
plt.show()
```

# True Label en columnas (Transpuesto)

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Ejemplo de etiquetas verdaderas y predicciones
y_true = [0, 1, 0, 1, 0, 1, 1]
y_pred = [0, 0, 0, 1, 0, 1, 1]

# Generar la matriz de confusi√≥n
cm = confusion_matrix(y_true, y_pred)

# Transponer la matriz para que True Label est√© en las columnas
cm_transposed = cm.T

# Crear la visualizaci√≥n con True Label en las columnas
disp = ConfusionMatrixDisplay(confusion_matrix=cm_transposed)
disp.plot(cmap='Blues')
plt.title('Matriz de Confusi√≥n (True Label en columnas)')
plt.show()
```

# True Label en Filas (Con etiquetas personalizadas)

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Ejemplo de etiquetas verdaderas y predicciones
y_true = [0, 1, 0, 1, 0, 1, 1]
y_pred = [0, 0, 0, 1, 0, 1, 1]

# Generar la matriz de confusi√≥n
cm = confusion_matrix(y_true, y_pred)

# Crear la visualizaci√≥n con True Label en las filas
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap='Blues')
plt.title('Matriz de Confusi√≥n (True Label en filas)')
plt.xlabel('Predicci√≥n')
plt.ylabel('Etiqueta Verdadera')  # Etiqueta personalizada para las filas
plt.show()
```

# True Label en Columnas (Con etiquetas personalizadas)

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Ejemplo de etiquetas verdaderas y predicciones
y_true = [0, 1, 0, 1, 0, 1, 1]
y_pred = [0, 0, 0, 1, 0, 1, 1]

# Generar la matriz de confusi√≥n
cm = confusion_matrix(y_true, y_pred)

# Transponer la matriz para que True Label est√© en las columnas
cm_transposed = cm.T

# Crear la visualizaci√≥n con True Label en las columnas
disp = ConfusionMatrixDisplay(confusion_matrix=cm_transposed)
disp.plot(cmap='Blues')
plt.title('Matriz de Confusi√≥n (True Label en columnas)')
plt.xlabel('Etiqueta Verdadera')  # Etiqueta personalizada para las columnas
plt.ylabel('Predicci√≥n')  # Etiqueta personalizada para las filas
plt.show()
```
